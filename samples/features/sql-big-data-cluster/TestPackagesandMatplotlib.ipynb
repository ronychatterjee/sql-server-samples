### Importing Matplotlib

Click again on the "Manage Packages" option in the Toolbar. This will open the Terminal window. In the Terminal window, you can find the instructions to install packages used by the python kernel through pip. To install "matplotlib" as a package, what you will type in the terminal is:

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
X = make_low_rank_matrix(20, 200, effective_rank=10)
total_variance = np.var(X, axis=0).sum()
```

Now let's show how the explained variance ratio estimated internally by a full PCA model can match the true explained variance ratio as measured from the data. This is an example on how we can use packages and plot the data using matplotlib for finding trends.

```python
pca = PCA().fit(X)
plt.plot(pca.explained_variance_ratio_, label='pca explained variance')

X_transformed = pca.transform(X)
variances = np.var(X_transformed, axis=0)
true_explained_variance_ratio = variances / total_variance

plt.plot(true_explained_variance_ratio, label='empirical explained variance')
plt.legend(loc='best')
plt.xlabel('pca components')
plt.ylabel('fraction of total variance')
```
